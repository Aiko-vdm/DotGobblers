```python
%load_ext autoreload
%autoreload 2

from agents.DotGobblers.config import PROJ_ROOT
from pathlib import Path
from agents.DotGobblers.analysis.GameAnalysis import GameAnalysis
import matplotlib.pyplot as plt
```

```python
exp_dir = Path(PROJ_ROOT, 'agents', 'DotGobblers', 'analysis', 'defense-agent-test', 'raw')

defensiveV1_json = Path(exp_dir, '14_3', 'contest_defensiveV1', 'scores', 'match_0.json' )
defensiveV2_json = Path(exp_dir, '14_3','contest_defensiveV2', 'scores', 'match_0.json' )

defensiveV1_exp = GameAnalysis(defensiveV1_json)
defensiveV2_exp = GameAnalysis(defensiveV2_json)
```
