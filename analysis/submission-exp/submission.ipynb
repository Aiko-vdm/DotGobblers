```python
%load_ext autoreload
%autoreload 2

from agents.DotGobblers.config import PROJ_ROOT
from pathlib import Path
from agents.DotGobblers.analysis.GameAnalysis import GameAnalysis
import matplotlib.pyplot as plt
```

```python
exp_dir = Path(PROJ_ROOT, 'agents', 'DotGobblers', 'analysis', 'submission-exp', 'raw', '17_3')

RefV3_vs_MV2_json = Path(exp_dir,'contest_ReflexV3-vs-MinimaxV2', 'scores', 'match_0.json')
MV2_NewDef_vs_MV2_json = Path(exp_dir, 'contest_Minimax_NewDefence-vs-MinimaxV2','scores', 'match_0.json')
NewOf_OldDef_vs_MV2_json = Path(exp_dir, 'contest_NewOffensive_OldDefensive-vs-MinimaxV2','scores', 'match_0.json')

RefV3_vs_MV2_exp = GameAnalysis(RefV3_vs_MV2_json)
MV2_NewDef_vs_MV2_exp = GameAnalysis(MV2_NewDef_vs_MV2_json)
NewOf_OldDef_vs_MV2_exp = GameAnalysis(NewOf_OldDef_vs_MV2_json)
```

```python
rows = 1
cols = 3

agg = plt.figure(figsize=[20, 5])
gs = agg.add_gridspec(rows, cols)

agg.suptitle(f'submission exp {RefV3_vs_MV2_exp.experiment_time}', fontsize=14)

# Create a variable to hold the last axis for sharing
ax1 = agg.add_subplot(gs[0, 0])  # Create first subplot to share y-axis
RefV3_vs_MV2_exp.generate_winplot()

ax2 = agg.add_subplot(gs[0, 1], sharey = ax1)
MV2_NewDef_vs_MV2_exp.generate_winplot()

ax3 = agg.add_subplot(gs[0, 2], sharey = ax1)
NewOf_OldDef_vs_MV2_exp.generate_winplot()
```
