```python
%load_ext autoreload
%autoreload 2

from agents.DotGobblers.config import PROJ_ROOT
from pathlib import Path
from agents.DotGobblers.analysis.GameAnalysis import GameAnalysis
import matplotlib.pyplot as plt
```

```python
exp_dir = Path(PROJ_ROOT, 'agents', 'DotGobblers', 'analysis', 'Agents-VS-Agents')

RV1_VS_RV2_json = Path(exp_dir, 'raw', '8_3', 'contest_ReflexV1-VS-ReflexV2', 'scores', 'match_0.json')
RV1_VS_MV1_json = Path(exp_dir, 'raw', '8_3', 'contest_ReflexV1-VS-MinimaxV1', 'scores', 'match_0.json')

RV1_VS_RV2_exp = GameAnalysis(RV1_VS_RV2_json)
RV1_VS_MV1_exp = GameAnalysis(RV1_VS_MV1_json)
```

```python
rows = 1
cols = 2

agg = plt.figure(figsize=[10, 5])
gs = agg.add_gridspec(rows, cols)

agg.suptitle(f'agents-vs-agents{RV1_VS_RV2_exp.experiment_time}', fontsize=14)

# Create a variable to hold the last axis for sharing
ax1 = agg.add_subplot(gs[0, 0])  # Create first subplot to share y-axis
RV1_VS_RV2_exp.generate_winplot()

ax2 = agg.add_subplot(gs[0, 1], sharey = ax1)
RV1_VS_MV1_exp.generate_winplot()
```
